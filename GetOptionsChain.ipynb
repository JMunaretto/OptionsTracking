```python
from yahoo_fin import options
from yahoo_fin.stock_info import *
```

```python
from matplotlib.pyplot import plot, figure
import matplotlib.pyplot as plt

import numpy as np
plt.style.use('ggplot')
plt.rcParams.update({'font.size': 20})

plt.rcParams['figure.figsize'] = 12, 6
plt.rcParams['axes.facecolor'] = (0.85, 0.85,0.85)

```

```python
tsla_dates= options.get_expiration_dates("tsla")

info = {}
for date in tsla_dates[:]:
    info[date] = options.get_options_chain("tsla", date)
    
ask = get_quote_data('TSLA')['ask']
calls = info[tsla_dates[1]]['calls']
puts = info[tsla_dates[1]]['puts']
```

```python
for aa in range(0, len(info)):
    calls = info[tsla_dates[aa]]['calls']
    puts = info[tsla_dates[aa]]['puts']   
    
    plt.plot((100 * calls['Strike'] / ask) - 100, 100*calls['Ask']/ ask, '.');plt.axis([0, 40, 0, 10])
    plt.xlabel('% Increase from Current Price')
    plt.ylabel('Premium %')
```
